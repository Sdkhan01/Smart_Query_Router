```python
##Finding the area of a rectangle 

from langgraph.graph import StateGraph , START , END
from typing import TypedDict
```

```python
class mystate(TypedDict):
    len : float
    width : float
    area : float
```

```python
#graph
graph = StateGraph(mystate)

def area(state : mystate) -> dict :
    len = state['len']
    width = state['width']
    state['area'] = len * width
    return state
```

```python
#Nodes
graph.add_node("area", area)
```

```python
graph.add_edge(START, "area")
graph.add_edge("area", END)
```

```python
graph.compile()
```

```python
comp = graph.compile()
```

```python
comp.invoke({"len":20,"width":10})
```
